```python
import pandas as pd
import time
```

```python
df = pd.DataFrame({'Time': ['212600', '173107', '110750', '162000', '0', '150222', '212500', '0', '61643', '0', '1151', '125154']})
```

```python
df
```

```python
def to_time(string):
    if len(string) == 6:
        string = [x for x in string]
        string.insert(2, " ")
        string.insert(5, " ")
        string = ''.join(string)
    elif len(string) == 4:
        string = [x for x in string]
        string.insert(2, " ")
        string.insert(5, " 00")
        string = ''.join(string)
    elif len(string) == 2:
        string = [x for x in string]
        string.insert(2, " 00 00")
        string = ''.join(string)
    else:
        string = "00 00 00"
    return string
```

```python
df['new_time'] = df.Time.apply(lambda x: to_time(x))
```

```python
df['the_time'] = pd.to_datetime(df['new_time'], format='%H %M %S').dt.time
```
